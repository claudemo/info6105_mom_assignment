```python
import pandas as pd
data = pd.read_csv('./data.csv', encoding='utf-8')
data.head()
```

```python
# Define the range of column titles
start_year = 65
end_year = 115

# Create a list of column titles to keep
columns_to_keep = [f'{start_year + i}Years' for i in range(end_year - start_year + 1)]

# Create a new DataFrame with the desired columns
filtered_data = data[columns_to_keep]
filtered_data.head()
```

```python
longevity = filtered_data.drop(0)

longevity.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have already read your data into the 'data' DataFrame
# If not, you can read the data from a CSV file like this:
# data = pd.read_csv('data.csv')

# Select the two rows you want to plot as separate datasets
data_row1 = longevity.loc[0]  # First row
data_row2 = longevity.iloc[1]  # Second row

# Create histograms for each row
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.hist(data_row1, alpha=0.5, label='Row 1', bins=10, edgecolor='k')
plt.hist(data_row2, alpha=0.5, label='Row 2', bins=10, edgecolor='k')

# Customize the plot
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Two Rows')
plt.legend()

# Display the plot
plt.show()
```
